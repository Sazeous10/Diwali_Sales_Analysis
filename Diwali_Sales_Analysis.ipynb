```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
df.info()
```

# Exploratory Data Analysis

```python
df.columns
```

### Product ID

```python
sales_product = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
Ps = sns.barplot(data = sales_product, x = 'Product_ID',y= 'Orders')
for bars in Ps.containers:
    Ps.bar_label(bars)
```

```python
amount_product = df.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
Pa = sns.barplot(data = amount_product, x = 'Product_ID',y= 'Amount')
for bars in Pa.containers:
    Pa.bar_label(bars)
```

*So maximum Sales are of the Prodcucts with Product Ids = P00265242 , P00110942 and Maximum Revenue is Generated by Product With Product ID = P00265242*

### Gender

```python
Gc = sns.countplot(x = 'Gender',data = df)

sns.set(rc={'figure.figsize':(6,5)})
for bars in Gc.containers:
    Gc.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)

sns.set(rc={'figure.figsize':(6,4)})
Gs = sns.barplot(x = 'Gender',y= 'Orders' ,data = sales_gen)
for bars in Gs.containers:
    Gs.bar_label(bars)
```

```python
amount_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

Ga = sns.barplot(x = 'Gender',y= 'Amount' ,data = amount_gen)
for bars in Ga.containers:
    Ga.bar_label(bars)
```

*From above graphs we can see that most of the buyers are females, most of the orders are from females and even the purchasing power of females is greater than men*

### Age

```python
Ac = sns.countplot(data = df, x = 'Age Group')
for bars in Ac.containers:
    Ac.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)

As = sns.barplot(x = 'Age Group',y= 'Orders' ,data = sales_age)
for bars in As.containers:
    As.bar_label(bars)
```

```python
amount_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
Am = sns.barplot(x = 'Age Group',y= 'Amount' ,data = amount_age)
for bars in Am.containers:
    Am.bar_label(bars)
```

*From above graphs we can see that most of the buyers and spenders are of age group between 26-35 yrs*

### State

```python
Sc = sns.countplot(data = df, x = 'State')
sns.set(rc={'figure.figsize':(30,10)})
for bars in Sc.containers:
    Sc.bar_label(bars)
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
Ss = sns.barplot(data = sales_state, x = 'State',y= 'Orders')
for bars in Ss.containers:
    Ss.bar_label(bars)
```

```python
amount_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
Sa = sns.barplot(data = amount_state, x = 'State',y= 'Amount')
for bars in Sa.containers:
    Sa.bar_label(bars)
```

*From above graphs customers, orders and amount spent are from Uttar Pradesh, Maharashtra and Karnataka respectively*

### Marital Status

```python
Mc = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(4,5)})
Mc.set_xticklabels(['Unmarried','Married'])
for bars in Mc.containers:
    Mc.bar_label(bars)
```

```python
df.groupby(['Marital_Status'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
```

```python
sales_MS = df.groupby(['Marital_Status'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)

sns.set(rc={'figure.figsize':(4,5)})
Ms = sns.barplot(data = sales_MS, x = 'Marital_Status',y= 'Orders')
Ms.set_xticklabels(['Unmarried','Married'])
for bars in Ms.containers:
    Ms.bar_label(bars)
```

```python
amount_MS = df.groupby(['Marital_Status'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
Ma = sns.barplot(data = amount_MS, x = 'Marital_Status',y= 'Amount')
Ma.set_xticklabels(['Unmarried','Married'])
for bars in Ma.containers:
    Ma.bar_label(bars)
```

*From above graphs we can see that most of the buyers are Unmarried and they have high purchasing power*

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
Oc = sns.countplot(data = df, x = 'Occupation')

for bars in Oc.containers:
    Oc.bar_label(bars)
```

```python
sales_oc = df.groupby(['Occupation'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
Os = sns.barplot(data = sales_oc, x = 'Occupation',y= 'Orders')
for bars in Os.containers:
    Os.bar_label(bars)
```

```python
amount_OC = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(30,10)})
Oa = sns.barplot(data = amount_OC, x = 'Occupation',y= 'Amount')

for bars in Oa.containers:
    Oa.bar_label(bars)
```

*From above graphs we can see that most of the buyers, orders and amount spent are from IT, Healthcare and Aviation sectors*

### Product Category

```python
sns.set(rc={'figure.figsize':(30,10)})
PCc = sns.countplot(data = df, x = 'Product_Category')

for bars in PCc.containers:
    PCc.bar_label(bars)
```

```python
sales_PC = df.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
PCs = sns.barplot(data = sales_PC, x = 'Product_Category',y= 'Orders')
for bars in PCs.containers:
    PCs.bar_label(bars)
```

```python
amount_PC = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
PCa = sns.barplot(data = amount_PC, x = 'Product_Category',y= 'Amount')
for bars in PCa.containers:
    PCa.bar_label(bars)
```

*From above graphs we can see that most of the sold products are from CLothing, Food & Electronics respectively and
max amount spent are on Food, Clothing & Electronics category*
